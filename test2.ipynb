```python
import pandas as pd

# Load the data
file_path = '2022_orders_discount_data.csv'
data = pd.read_csv(file_path)

# Step 1.1: Inspect the data
print("First few rows of the data:")
print(data.head())

print("\nSummary of the data:")
print(data.info())

# Step 1.2: Check for missing values
missing_values = data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# Step 1.3: Convert date_of_order to datetime format
data['date_of_order'] = pd.to_datetime(data['date_of_order'])

# Step 1.4: Check for duplicates
data = data.drop_duplicates()

print("\nShape of the data after removing duplicates:")
print(data.shape)

# Step 1.5: Descriptive statistics
print("\nDescriptive statistics:")
print(data.describe())

# Step 1.6: Check for negative or zero orders
negative_orders = data[data['orders'] < 0]
print("\nNegative orders:")
print(negative_orders)

# Check for unusually high discounts (above 100%)
high_discounts = data[data['discount'] > 100]
print("\nUnusually high discounts (above 100%):")
print(high_discounts)

# Step 1.7: Summary of logical inconsistencies
if negative_orders.empty:
    print("\nNo negative orders found.")
else:
    print("\nNegative orders detected. Review and handle these values.")

if high_discounts.empty:
    print("No unusually high discounts found.")
else:
    print("Unusually high discounts detected. Review and handle these values.")

```
